```
def f(x):
  Z = 24.44321494051954
  if abs(x) > 7:
    return 0
  elif abs(x) > 3:
    return 3*(1-(x/7)**2)**0.5/Z
  elif abs(x) > 1:
    return ((3-abs(x))/2-3/7*10**0.5*((3-x**2+2*abs(x))**0.5-2))/Z
  elif abs(x) > 0.75:
    return (9 - 8 * abs(x)) / Z
  elif abs(x) > 0.5:
    return (3 * abs(x) + 0.75) / Z
  else:
    return 2.25 / Z
```
